## Adult.csv

```python
dataset = "kc1"
url = '/content/drive/MyDrive/Luiz Gabriel - PC/13.Pós Data Science UFPR/00.TCC/Especializacao_Data_Science_Big_Data_Monografia/data/kc1.csv'
target = 'defects'
df = pd.read_csv(url, sep=',')
df.head()
```

```python
df_train = df.copy()                                                # Create copy of pandas DataFrame
df_train[target] = df_train[target].map({True: 1, False: 0})  # Replace boolean by data_new1
```

```python
# treino por 1 hora no máximo
def train_model(df, target):
    X_train, X_test, y_train, y_test = train_test_split(df.loc[:, df.columns != target], df[target], test_size=0.30, random_state=123)
    cls = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=60,
                                                           seed = 123,
                                                           scoring_functions=[accuracy,log_loss,roc_auc])
    cls.fit(X_train, y_train)
    return cls

def get_metric_result(cv_results):
    results = pd.DataFrame.from_dict(cv_results)
    results = results[results['status'] == "Success"]
    cols = ['rank_test_scores', 'param_classifier:__choice__', 'mean_test_score','mean_fit_time']
    cols.extend([key for key in cv_results.keys() if key.startswith('metric_')])
    results = results[cols]

    return results

def get_leaderboard():
    leaderboard = cls.leaderboard(ensemble_only=True)
    leaderboard.reset_index(drop=True, inplace=True)
    return leaderboard

def merge_leaderbord_results(x,y):
    full = pd.merge(x,y, how='inner', left_on=['type','duration'], right_on=['param_classifier:__choice__','mean_fit_time'])
    full["g"] = "g"
    return full

def grouped_weighted_avg(values, weights, by):
    return (values * weights).groupby(by).sum() / weights.groupby(by).sum()

def metrics_full():
    metricas = pd.DataFrame()
    metricas["acc"] = grouped_weighted_avg(values=full.metric_accuracy, weights=full.ensemble_weight, by=full.g)
    metricas["log_loss"] = grouped_weighted_avg(values=full.metric_log_loss, weights=full.ensemble_weight, by=full.g)
    metricas["roc_auc"] = grouped_weighted_avg(values=full.metric_roc_auc, weights=full.ensemble_weight, by=full.g)
    metricas["dataset"] = dataset
    return metricas
```

```python
start_time = datetime.now()

cls = train_model(df= df_train, target= target)
results_ = get_metric_result(cv_results = cls.cv_results_)
leaderboard_ = get_leaderboard()
full = merge_leaderbord_results(leaderboard_, results_)
metrics = metrics_full()

end_time = datetime.now()
execution_time = end_time - start_time

metrics["execution_time"] = execution_time
metrics
#run_it_all()


```

# - Auto-Sklearn

```python
model = AutoSklearnClassifier(time_left_for_this_task=5*60, per_run_time_limit=30, n_jobs=8)
# perform the search
model.fit(X_train, y_train)
# summarize
print(model.sprint_statistics())
# evaluate best model
y_hat = model.predict(X_test)
acc = accuracy_score(y_test, y_hat)
print("Accuracy: %.3f" % acc)
```

```python
from sklearn.metrics import classification_report
target_names = ['Benigno', 'Maligno']
print(classification_report(y_test, y_hat, target_names=target_names))
```
