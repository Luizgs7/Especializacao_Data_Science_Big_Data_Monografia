## Adult.csv

```python
dataset = "adult"
url = '/content/drive/MyDrive/Luiz Gabriel - PC/13.Pos Data Science UFPR/00.TCC/Especializacao_Data_Science_Big_Data_Monografia/data/adult.csv'
df = pd.read_csv(url, sep=',')
df.head()
```

```python
start_time = datetime.now()

# Editing pycaret's setup
sp = setup(data=df,
           target='class',
           silent = True,
           session_id=123)

# Add logloss mtric
from sklearn.metrics import log_loss
add_metric('logloss', 'Log Loss', log_loss, greater_is_better=False, target="pred_proba")

# Compare models and choose de best by AUC
best_model = compare_models(sort="auc") # sort='recall'
model = pull()["Model"][0]

# Tuning best model hyperparameters
best_model_tuned = tune_model(best_model)

end_time = datetime.now()
execution_time = end_time - start_time
print('Duration: {}'.format(end_time - start_time))

# Getting metrics
metrics = pull()[["Accuracy", "Log Loss","AUC"]]
metrics = metrics.filter(items = ['Mean'], axis=0)
metrics["execution_time"] = execution_time
metrics["dataset"] = dataset
metrics["modelo"] = model
```

```python
# Create model_metrics dataframe to store all the experiments metrics
models_metrics = metrics.copy()


del sp
del best_model
del best_model_tuned
```
